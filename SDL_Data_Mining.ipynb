```
class SDL:
  def __init__(self, index, statements, link):
    self.statements = statements
    self.index = index
    self.videoLink = link

    self.actors = []
    self.scene = []
    self.actions = []  

  def getActors(self, statement):
    '''
    *car
    *bus
    *truck
    *cyclists
    *pedestrian
    '''
    phrase = statement
    if ('car' in phrase):
      self.actors.append('car')
    if ('bus' in phrase):
      self.actors.append('bus')
    if ('truck' in phrase):
      self.actors.append('truck')
    if ('cyclist' in phrase):
      self.actors.append('cyclist')
    if ('pedestrain' in phrase):
      self.actors.append('pedestrain')
  

  def getScene(self, statement):
    '''
    *intersection
    *number of lanes
    *crosswalk
    *bridge
    *traffic light
    *stop sign
    *u-turn
    '''

    phrase = statement
    if ('intersection' in phrase):
      self.scene.append('intersection')
    if ('crosswalk' in phrase):
      self.scene.append('crosswalk')
    if ('bridge' in phrase):
      self.scene.append('bridge')
    if ('traffic light' in phrase):
      self.scene.append('traffic light')
    if ('light' in phrase):
      self.scene.append('traffic light')
    if ('stop sign' in phrase):
      self.scene.append('stop sign')
    if ('u-turn' in phrase):
      self.scene.append('u-turn')

  def getActions(self, statement): 
    '''
    *Turn
    *Merge
    *Accelerate
    *brake
    *stop
    *slow down
    *Moving
    *walking
    *crossing
    *Parking
    *forward
    *change lane
    '''
    phrase = statement
    if ('turn' in phrase):
      self.actions.append('turn')
    if ('merge' in phrase):
      self.actions.append('merge')
    if ('accelerate' in phrase):
      self.actions.append('accelerate')
    if ('brake' in phrase):
      self.actions.append('brake')
    if ('stop' in phrase):
      self.actions.append('stop')
    if ('slow' in phrase):
      self.actions.append('slow')
    if ('moving' in phrase):
      self.actions.append('moving')
    if ('walking' in phrase):
      self.actions.append('walking')
    if ('cross' in phrase):
      self.actions.append('cross')
    if ('parking' in phrase):
      self.actions.append('parking')
    if ('forward' in phrase):
      self.actions.append('forward')
    if ('change' in phrase):
      self.actions.append('change lanes')
```
