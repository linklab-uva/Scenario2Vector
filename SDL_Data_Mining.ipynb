```python
class Actor:
    def __init__(self, description):
        """
        self.description should be one of the following:
          ego
          light vehicle
          heavy vehicle
          cyclist
          pedestrian
        """
        self.description = description
        self.action = ""
```

```python
# Aron's Version
class SDL:
  def __init__(self, index, statements, link):
    self.statements = statements
    self.index = index
    self.videoLink = link
    self.stemmer = nltk.stem.PorterStemmer()
    
    self.actor_list = {'car':'light vehicle',
                       'bus':'heavy vehicle',
                       'truck':'heavy vehicle',
                       'cyclist':'cyclist',
                       'pedestrian':'pedestrian',
                       'ambulance':'heavy vehicle',
                       'minivan':'light vehicle'
                      }
    self.scene_list = ['intersection','crosswalk','bridge','light',
                  'stop sign','u-turn',]#,'traffic light']

    init_action_list = ['turn','merge','accelerate','brake','stop','slow',
                        'move','walk','cross','park','forward']#,'change lane']
    self.action_list = []
    for a in init_action_list:
        self.action_list.append(self.stemmer.stem(a))

    self.actors = []
    self.scene = []
    
  def getDescriptors(self, statement):    
    phrase = statement.lower().split(" ")
    for word in phrase:
        if (word in self.actor_list):
            self.actors.append(Actor(self.actor_list[word]))
        elif (word in self.scene_list):
            self.scene.append(word)
        elif (self.stemmer.stem(word) in self.action_list):
            if len(self.actors) > 0 and self.actors[-1].action == "":
                self.actors[-1].action = word
```
